```python
import time
# Threading example
import threading
def thread_function():
    for i in range(50):
        time.sleep(0.2)
        print(i)

def thread_function2():
    for i in range(50):
        time.sleep(0.2)
        print("\n" + str(i))
        
        
x = threading.Thread(target=thread_function2, args=())
y = threading.Thread(target=thread_function, args=())
x.start()
y.start()
```
