```python
driver = webdriver.Chrome(r'./chromedriver')
driver.get("https://web.whatsapp.com/")
wait = WebDriverWait(driver, 600)
```

```python
chat = driver.find_element_by_class_name("_1ays2")
chat = chat.find_elements_by_xpath("div")
```

```python
# Chat loop to check new messege
while True:
    # Saving the last messege
    lastSavedMessege = chat[len(chat)-1]
    
    # Chat div
    chat = driver.find_element_by_class_name("_1ays2")
    chat = chat.find_elements_by_xpath("div")
    
    # Checking for new messege
    lastMessege = chat[len(chat)-1]
    if lastMessege != lastSavedMessege:
        if MessegeChecker(lastMessege):
             print(MessegeChecker(lastMessege))
```

```python
def MessegeChecker(lastMessege):
    # Check if the element is a messege by checking if the first element inside is an emty span 
    # Change the statement if whatsapp changes the stracture
    # Xpath search for a class contains 'messege-in' to verify it's a messege: 
    
    try:
        if 'message-in' not in lastMessege.get_attribute("class"):
            lastMessege.find_element_by_xpath("*[contains(@class,'message-in')]")
            
        imageUrl = None
        
        # It's a messege
        # Let's check what type of messege is it
        try:
            # Checking if the message has metadata inside the data-pre-plain-text attribute in the message div
            MessageMeta = lastMessege.find_element_by_xpath(".//*[@data-pre-plain-text]").get_attribute("data-pre-plain-text")

        except:
            # That means the message div has no metadata and we should scrape it mannually
            MessageMeta = None


        if MessageMeta != None:
            
            try:
                
                # Seperate the metadata values into variables: name, date, time.
                MessageMeta = MessageMeta.replace("[", "")
                MessageMeta = MessageMeta.replace("]", "")
                MessageMeta = MessageMeta.replace(",", "")
                MessageMeta = MessageMeta.split(" ")
                
                date = MessageMeta[1]
                time = MessageMeta[0]
                
                name = (' '.join(MessageMeta[2:])).replace(":", "")[0:-1]
                
                name = name.replace("\u202c", "")
                name = name.replace("\u202a", "")
                
            except:
                MessageMeta == None

        elif MessageMeta == None:
            
            # Scraping message metadata manually
            name = lastMessege.text.split("\n")[0] + " "
            time = lastMessege.text.split("\n")[-1]
            date = None
            
        # Checking if it's image by checking if the image path exists
        try:
            image = lastMessege.find_element_by_xpath(".//img")
            if "blob:https://web.whatsapp.com" in image.get_attribute("src"):
                # It's an image
                messageType = "image"

                # Checking the text from the message with a way that recognize if the image has also text
                if len(lastMessege.text.split("\n")) > 3:
                    content = lastMessege.text.split("\n")[-2]
                else:
                    content = None
                
                # Now we will take the image url
                imageUrl = image.get_attribute("src")

            else:
                # Not image - can be an emoji, location, etc...
                # Let's check if the message is an emoji: 
                # Emoji's img has data-plain-text attribute with a value of an emoji.

                if image.get_attribute("data-plain-text") == None:
                    # Not emoji
                    # Go to exception
                    1/0

                else:
                    # It's an emoji, We can try to get the emoji code
                    messageType = "emoji"

                    try:
                        content = image.get_attribute("data-plain-text")

                    except:
                        None

        except:


            messageType = "text"

            # Statement to check if the message is a contact
            try:
                # Contact
                lastMessege.find_element_by_xpath(".//span[@data-icon='default-user']")
                messageType = "contact"

            except:
                None

            try:
                lastMessege.find_element_by_xpath(".//span[@data-icon='audio-download']")
                # file message 
                messageType = "file"

            except:
                None

            try:
                # Searching for the little microphone logo means it's a voice message
                lastMessege.find_element_by_xpath(".//span[contains(@data-icon,'ptt')]")
                messageType = "audio"

            except:
                None
            
            try:
                content = lastMessege.find_element_by_xpath(".//*[@data-pre-plain-text]").find_element_by_xpath(".//span").text
            
            except:
                content = lastMessege.text.split("\n")[-2]

        if content == "הודעה זו נמחקה" and date == None:
            messageType = "DeletedMessage"

        
        messageData = {"Name": name, "Date": date, "Time": time, "Type": messageType, "Content": content, "File Url": imageUrl}
        return messageData

    except:
        # not an in message
        # print("Not a message")
        False

```

```python
import time
# Threading example
import threading
def thread_function():
    for i in range(50):
        time.sleep(0.2)
        print(i)

def thread_function2():
    for i in range(50):
        time.sleep(0.2)
        print("\n" + str(i))
        
        
x = threading.Thread(target=thread_function2, args=())
y = threading.Thread(target=thread_function, args=())
x.start()
y.start()
```

```python
# Send a message
def send_message(message):
    try:
        try:
            
            # Finding the second input of the page - which is the input to send a message
            driver.find_elements_by_tag_name("input")[1].send_keys(message)
            time.sleep(0.1)
        
        except:
            # input box to send message class = _13mgZ
            textTypeField = driver.find_element_by_class_name('_13mgZ')
            for div in textTypeField.find_elements_by_xpath(".//div"):
                try:
                    div.send_keys(message)
                except:
                    None

       # Finding the button to send by searching a spane whith the send icon
        send = driver.find_element_by_xpath(".//span[@data-icon='send']")
        send.click()

    except Exception as e:
        print(e)
```

```python
def contactMeta(contact):
    # Green new message circle span class = 'P6z4j'
    try:
        newMessageNumber = contact.find_element_by_class_name("P6z4j").span
        
    except:
        newMessageNumber = None
        
    contactName = contact.find_element_by_xpath(".//span[@title]").text
    groupText = contact.find_element_by_xpath(".//span/following-sibling::span").text
    if groupText != "" or groupText != None:
        # It's a group
        return {"name": contactName, "text": groupText, "New Message Number": newMessageNumber}
```

```python
# Contacts loop to check new messege
lastContact = None
while True:
    # Saving the last contact
    lastSavedContact = lastContact
            
    for contact in driver.find_elements_by_class_name('X7YrQ'):
        # Catch always the fist contact by getting the placement css
        if 'translateY(0px);' in contact.get_attribute("style"):
            lastContact = contact
            
    
    # Checking for new messege
    if lastContact != lastSavedContact:
        print("message")
        # New message
        contactDict = contactMeta(lastContact)
        print(contactDict)
        if contactDict["text"] == "מזל טוב":
            lastContact.click()
            time.sleep(0.1)
            send_message("מזל טוב!!!")
```

```python
def scroll_to_date(date, day_before=None):

    driver.find_element_by_tag_name("body").click()
    while True:

        Chatdate = driver.find_element_by_class_name("copyable-area").find_element_by_tag_name("span").text
        
        if Chatdate == date:
            break
            
        if date == day_before:
            print("Date not found.")
            break


        driver.find_element_by_tag_name("body").send_keys(Keys.ARROW_UP * 30)

```
